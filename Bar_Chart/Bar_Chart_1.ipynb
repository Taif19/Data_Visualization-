```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
from matplotlib import style
```

```python
material_data = {'fiber 1':100, 'fiber 2':90, 'fiber 3':85, 'fiber 4':112}
material = list(material_data.keys())
cmp = list(material_data.values())
```

```python
plt.bar(material,cmp, color="g", width=0.5)
plt.xlabel('Name of material')
plt.ylabel('MPa of Material')
plt.title('Composite Data')
```

```python
color = ['royalblue', 'red', 'purple', 'yellow']
plt.bar(material,cmp, color=color, width=0.5)
plt.xlabel('Name of material')
plt.ylabel('MPa of Material')
plt.title('Composite Data')
```

```python
fb = (120, 90, 110, 100, 95)
yt = (80, 90, 75, 85, 95)
ig = (50, 40, 55, 35, 45)
sc = (20, 30, 35, 25, 28)
```

```python
xpos = np.arange(len(fb))
xpos
```

```python
style.use('ggplot')
barWidth = 0.2
plt.bar(xpos, fb, color='royalblue', width = barWidth, label='FB')
plt.bar(xpos+0.2, yt, color='red', width = barWidth, label='YT')
plt.bar(xpos+0.4, ig, color='purple', width = barWidth, label='IG')
plt.bar(xpos+0.6, sc, color='yellow', width = barWidth, label='SC')
plt.xticks(xpos+0.3, ('Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'))
plt.xlabel('Day')
plt.ylabel('Time spend in minutes')
plt.title('Time spend on social media')
plt.legend()
```

```python
fiber_1 = (120, 100, 110, 100, 115)
fiber_2 = (80, 100, 70, 85, 90)
fiber_3 = (50, 40, 35, 55, 49)
x = ['Test-1', 'Test-2', 'Test-3', 'Test-4', 'Test-5']
xpos = np.arange(len(x))
barWidth = 0.4
plt.figure(figsize=(10,7))
plt.bar(xpos, fiber_1, color='royalblue', width= barWidth, label='FBR 3')
plt.bar(xpos, fiber_2, bottom=fiber_1, color='red', width= barWidth, label='FBR 1')
plt.bar(xpos, fiber_3, bottom=fiber_2, color='purple', width= barWidth, label='FBR 2')
plt.xlabel('Sample no')
plt.ylabel('MPa')
plt.title('Tensile Test')
plt.xticks(xpos, x)
plt.legend()
```
