```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import projections
from mpl_toolkits.mplot3d import Axes3D
```

```python
def f(x,y):
    return x**2+y**2
x=np.arange(-5, 5,0.1)
y=np.arange(-5, 5,0.1)

X, Y= np.meshgrid(x,y)
Z=f(X,Y)
fig=plt.figure()
ax =fig.add_subplot(projection='3d')
surf=ax.plot_surface(X,Y,Z)
```

```python
from matplotlib import cm
def f(x,y):
    return x**2+y**2
x=np.arange(-5, 5,0.1)    #defining array for x plot 
y=np.arange(-5, 5,0.1)    #defining array for x plot

X, Y= np.meshgrid(x,y)
Z=f(X,Y)
fig=plt.figure()
ax =fig.add_subplot(projection='3d')
surf=ax.plot_surface(X,Y,Z, cmap=cm.cool)
ax.set_xlabel('x values')
ax.set_ylabel('y values')
ax.set_zlabel('z value') 
ax.set_zlim(0.100)
plt.colorbar(surf)
```

```python
from matplotlib import cm
def f(x,y):
    return x**2+y**2
x=np.arange(-5, 5,0.1)
y=np.arange(-5, 5,0.1)

X, Y= np.meshgrid(x,y)
Z=f(X,Y)
fig=plt.figure(figsize=(7,5))
ax =fig.add_subplot(projection='3d')
surf=ax.plot_surface(X,Y,Z, cmap=cm.gist_rainbow)
ax.set_xlabel('x values')
ax.set_ylabel('y values')
plt.colorbar(surf)
ax.set_zlim(0,70)
ax.set_zlabel("Z value") 
ax.zaxis.labelpad=-0.7              #z label fix in jn
ax.set_box_aspect(None, zoom=0.95)  #z label fix in jn
```
