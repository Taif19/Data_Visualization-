```python
import matplotlib.pyplot as plt
```

```python
x=[30,14,5,17,3,7,8,25]
y=[6,8,25,7,10,16,6,20]
```

```python
plt.scatter(x,y, s=300, color='y')
plt.title("Test Plot", fontsize=20)
plt.xlabel("x data", fontsize=15)
plt.ylabel("y data", fontsize=15)
plt.show()
```

```python
color=["r","y","b","g","b","r","y","b"]
plt.scatter(x,y, s=300 ,color=color)
plt.title("Test Plot", fontsize=20)
plt.xlabel("x data", fontsize=15)
plt.ylabel("y data", fontsize=15)
plt.show()
```

```python
sizes=[400,200,100,300,450,230,150,150]
from matplotlib import colormaps
color=[30,29,56,40,55,12,60,10]
plt.xlim(0, 50)
plt.ylim(0, 30)
plt.scatter(x,y, s=sizes, c=color, cmap="viridis", marker= "v")
plt.colorbar()
plt.title("Test Plot", fontsize=20, loc="right")
plt.xlabel("x data", fontsize=15)
plt.ylabel("y data", fontsize=15)
plt.show()
```

```python
from matplotlib import colormaps
x1=[3,15,20,12,17,2,19,22]
y1=[5,6,3,22,17,10,20,16]
sizes=[400,200,100,300,450,230,150,150]
color=[30,29,56,40,55,12,60,10]
plt.xlim(0, 50)
plt.ylim(0, 30)
plt.scatter(x,y, s=sizes, c=color, marker= "^",cmap="viridis")
plt.scatter(x1,y1, s=sizes, c=color,cmap="viridis", marker= "o")
plt.colorbar()
plt.title("Test Plot", fontsize=20,)
plt.xlabel("x data", fontsize=15)
plt.ylabel("y data", fontsize=15)
plt.show()
```

```python
x1=[3,15,20,12,17,2,19,22]
y1=[5,6,3,22,17,10,20,16]
plt.xlim(0, 35)
plt.ylim(0, 35)
plt.scatter(x,y, s=100, c="r", marker= "x", label='test_cross_marked_data')
plt.scatter(x1,y1, s=100, c="b", marker= "o", label='test_color_marked_data')
plt.title("Test Plot", fontsize=20,)
plt.xlabel("x data", fontsize=10)
plt.ylabel("y data", fontsize=10)
plt.grid()
plt.legend()
plt.show()
```
